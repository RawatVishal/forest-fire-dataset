# Numerical Analysis

```python
#separating numerical and categorical data
num_data = [i for i in df1.columns if df1[i].dtype != 'O']
cat_data = [i for i in df1.columns if df1[i].dtype=='O']

#printing number of numerical and categorical data
print('we have {} numerical data which is {} '.format(len(num_data),num_data))
print('\nwe have {} categorical data which is {}'.format(len(cat_data),cat_data))
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of numerical features',fontsize=20, fontweight='bold', alpha=0.8, y=1)
for i in range(0,len(num_data)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df1[num_data[i]],shade=True, color='b')
    plt.xlabel(num_data[i])
    plt.tight_layout()
```

```python
# There is no graph in year section because year is only 2012 so there will be no graph
#Rain,FWI,ISI,BUI,DC,DMC are right skewed or positively skewed
#FFMC  is negatively skewed or left skewed
#FWI,ISI,BUI,DC,DMC,FFMC contain outliers.
```

```python
plt.figure(figsize=(20,15))
plt.suptitle('Univariate Analysis of categorical features',fontsize=20,fontweight='bold',alpha=0.8,y=1)
for i in range(0,len(cat_data)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df1[cat_data[i]])
    plt.xlabel(cat_data[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

```python
# from above graph we can see that fire count higher thar no fire count 
```
